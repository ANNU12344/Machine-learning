```python
def read_text_from_file(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    return text

# Example usage:
file_path = '/content/TextRank.txt'  # Replace 'example.txt' with the path to your text file
text = read_text_from_file(file_path)
print(text)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
sentences = sent_tokenize(text)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
english_stopwords = stopwords.words('english')

# Print the list of English stopwords
print(english_stopwords)
```

```python
word_tokens = [word_tokenize(sentence) for sentence in sentences]
```

```python


def remove_stopwords_from_sentences(sentences):
    stop_words = set(stopwords.words('english'))
    filtered_sentences = []
    for sentence in sentences:
        filtered_sentence = [word for word in sentence if word.lower() not in stop_words]
        filtered_sentences.append(filtered_sentence)
    return filtered_sentences

# Example usage:


filtered_sentences = remove_stopwords_from_sentences(word_tokens)
for sentence in filtered_sentences:
    print(sentence)
```

```python
import numpy as np

# Load pre-trained GloVe word embeddings
embeddings_index = {}
with open('/content/glove.6B.100d.txt', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs

# Function to get the GloVe embedding vector for a word
def get_embedding(word):
    return embeddings_index.get(word, np.zeros(100))  # Return zero vector if word not found

# Apply GloVe embeddings to filtered sentences
glove_embeddings = []
for sentence in filtered_sentences:
    sentence_embeddings = [get_embedding(word) for word in sentence]
    glove_embeddings.append(sentence_embeddings)

# Now glove_embeddings contains the GloVe embeddings for the words in your filtered sentences
# Each element in glove_embeddings is a list of GloVe embedding vectors for a sentence

# Example usage:
for embeddings in glove_embeddings:
    print(embeddings)
```
